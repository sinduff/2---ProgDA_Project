```python
#import python standard libaries 
import numpy as np
from numpy.random import default_rng
rng = default_rng()

import csv
import pandas as pd
%matplotlib inline
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_style("darkgrid")
```

Variety 1 has 65
Variety 2 has 60
variety 3 has 53


```python
#Code Cell 5 - Synthesising the Data for variety 1

#Setting a seed to stop the data changing everytime that it runs.  
rng = default_rng(100)

#Step 1  ..create individual lists of random values 
#.. create lists is a more efficient use of memory than creating 
#and populating a dataframe - ref point 21

#Create random integer values between specified ranges using rng.integers() and rng.uniform()
variety1 = [rng.integers(1,2) for i in range(65)]
magnesium1 = [rng.integers(80,150) for i in range(65)]
proline1 = [rng.integers(300,1000) for i in range(65)]
alcohol1 = [rng.uniform(10.0,15.0)for i in range(65)]
ash1 = [rng.uniform(2.0,4.0)for i in range(65)]
flavanoids1 = [rng.uniform(1.0,6.0)for i in range(65)]
hue1 = [rng.uniform(0.50,2.0)for i in range(65)]
malicAcid1 = [rng.uniform(2.0,7.0)for i in range(65)]
alcalintyOfAsh1 = [rng.uniform(10,50)for i in range(65)]
totalPhenols1 = [rng.uniform(1.5,5.0)for i in range(65)]
nonFlavanoidsPhenols1 = [rng.uniform(0.0,1.0)for i in range(65)]
proanthocyanis1 = [rng.uniform(1.0,4.0)for i in range(65)]
colorIntensity1 = [rng.uniform(10.0,15.0)for i in range(65)]
diluted1 = [rng.uniform(1.0,5.0)for i in range(65)]

#Step 2 ... create a dataframe in pandas, and name the columns
df1 = pd.DataFrame(list(zip(variety1, alcohol1, malicAcid1, ash1, alcalintyOfAsh1, magnesium1, totalPhenols1, flavanoids1, nonFlavanoidsPhenols1, proanthocyanis1, colorIntensity1, hue1, diluted1, proline1)))
df1.columns=['Variety', 'Alcohol', 'Malic Acid','Ash','Alcanlinity of Ash','Magnesium','Total Phenols','Flavanoids','NonFlavanoids Phenols','Proanthocyanis','Color Intensity', 'Hue', 'Diluted','Proline' ]
```
